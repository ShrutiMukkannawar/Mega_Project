```
import numpy as np
#from keras.models import load_model
#from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
```

```
model_best = load_model('Final_Dataset/best_model_3class.hdf5',compile = False)
```

```
Apple_list = ['Mature','Premature','Rotten']
```

```

def predict_class(model, images, show = True):
  for img in images:
    img = image.load_img(img, target_size=(299, 299))
    img = image.img_to_array(img)                    
    img = np.expand_dims(img, axis=0)         
    img /= 255.                                      

    pred = model.predict(img)
    index = np.argmax(pred)
    
    pred_value = Apple_list[index]
    if show:
        plt.imshow(img[0])                           
        plt.axis('off')
        plt.title(pred_value)
        plt.show()
```

```

images = []

images.append('Final_Dataset/NewImages/D (1).jpg')
images.append('Final_Dataset/NewImages/D (2).jpg')
images.append('Final_Dataset/NewImages/D (3).jpg')
images.append('Final_Dataset/NewImages/D (17).jpg')
images.append('Final_Dataset/NewImages/D (6).jpg')
images.append('Final_Dataset/NewImages/D (8).jpg')
images.append('Final_Dataset/NewImages/D (10).jpg')
images.append('Final_Dataset/NewImages/D (23).jpg')
images.append('Final_Dataset/NewImages/D (25).jpg')
images.append('Final_Dataset/NewImages/D (31).jpg')
images.append('Final_Dataset/NewImages/D (32).jpg')


predict_class(model_best, images, True)
```
